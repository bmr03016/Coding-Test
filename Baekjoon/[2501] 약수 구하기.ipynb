```python
N, k = map(int,input().split())
divisor = []
try:
    for i in range(1, N+1): # i = 1, 2, 3, 4, 5, 6
        if N % i == 0: # 나머지가 0이면 
            divisor.append(i) # 약수 리스트에 추가 
                                # [ 1, 2, 3, 6 ]
    print(divisor[k-1]) # 3
    
except:
    print(0)   
```

```python
N, k = map(int,input().split())
divisor = []
try:
    for i in range(1, N+1): # i = 1, 2, 3, 4, 5, 6
        if N % i == 0: # 나머지가 0이면 
            divisor.append(i) # 약수 리스트에 추가 
                                # [ 1, 2, 3, 6 ]
    print(divisor[k-1]) # 3
    
except:
    print(0)   
```

'N의 약수의 개수가 K개보다 적어서 K번째 약수가 존재하지 않을 때'를 어떻게 쓰지하다가 예외 구문을 생각했고 try - except 식으로 구현했는데 다른 분들의 풀이를 찾아보니 이렇게 푸신 분이 거의 없었다..

```python
N, K = map(int, input().split())
divisor = []

for i in range(1, N+1):
    if (N % i == 0):
        divisor.append(i)
        
if len(divisor) < K:
    print(0)
else:
    print(divisor[K-1])
```

그냥 if - else로 많이들 풀이하셨더라..
