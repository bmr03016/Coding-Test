```python
X= int(input())
N= int(input())

sum = 0
for _ in range(N):
    a,b = map(int, input().split())
    sum += a*b

if sum == X:
    print("Yes")
else:
    print("No")
```

원래는 X,N = map(int, input().split())으로 한꺼번에 입력받으려했는데 런타임 에러가 뜨길래 따로 입력받게 수정했다! 쉬웠던 문제..
