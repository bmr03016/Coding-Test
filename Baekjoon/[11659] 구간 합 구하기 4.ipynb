# 👆 누적합

- 전체 수열의 누적합을 구하면 부분 순열의 합을 빠르게 구할 수 있다.
- i번째부터 j번째까지 부분합은 "[ j ]번째 까지 누적합 - [i - 1]번째 까지 누적합"으로 빠르게 구할 수 있다.

```python
N, M = map(int, input().split()) # 5 3
arr = [0] + list(map(int, input().split())) # 계산 시 인덱스를 초과하지 않게하기 위해 0추가
                                            # [0, 5, 4, 3, 2, 1]

for n in range(1, N + 1): # n = 1, 2, 3, 4, 5 
    arr[n] = arr[n - 1] + arr[n] # arr[1] = arr[0] + arr[1]
    
for _ in range(M): # M번 반복
    i, j = map(int, input().split()) # 2 4
    print(arr[j] - arr[i - 1]) # arr[4] - arr[1] = (5 + 4 + 3 + 2 + 0) - (5 + 0) = 9
```
