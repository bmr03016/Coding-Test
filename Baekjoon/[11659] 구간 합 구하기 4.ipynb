```python
N, M = map(int, input().split()) # 5 3
arr = [0] + list(map(int, input().split())) # ê³„ì‚° ì‹œ ì¸ë±ìŠ¤ë¥¼ ì´ˆê³¼í•˜ì§€ ì•Šê²Œí•˜ê¸° ìœ„í•´ 0ì¶”ê°€
                                            # [0, 5, 4, 3, 2, 1]

for n in range(1, N + 1): # n = 1, 2, 3, 4, 5 
    arr[n] = arr[n - 1] + arr[n] # arr[1] = arr[0] + arr[1]
    
for _ in range(M): # Më²ˆ ë°˜ë³µ
    i, j = map(int, input().split()) # 2 4
    print(arr[j] - arr[i - 1]) # arr[4] - arr[1] = (5 + 4 + 3 + 2 + 0) - (5 + 0) = 9
```
