```python
S = input() # baekjoon 
alpha = 'abcdefghijklmnopqrstuvwxyz' 
count_list = [] # 알파벳 갯수를 세줄 리스트

for i in range(len(alpha)): # i = 0, 1, 2, ...25
    if alpha[i] in S: # a, b, c...가 baekjoon에 있다면
        count_list.append(S.count(alpha[i])) # a, b, c의 갯수를 append하고
    else: # 없는 원소는
        count_list.append(0) # 0을 append
    
print(*count_list) 
```

🌟 **해당하는 알파벳 공간 찾는법**은 아스키코드를 활용해서도 가능하다!
- alphabet 갯수만큼 리스트 공간을 만들어주고 [0]
- 이때, s는 b, a, e, k, j, o, o, n
- lst의 [b의 아스키코드 98에서 97 빼기]= += 1
- lst[1] += 1 -> b자리에 1을 더해준다

```python
lst =[0]*26
for s in input(): 
    lst[ord(s)-97] += 1
print(*lst)
```
