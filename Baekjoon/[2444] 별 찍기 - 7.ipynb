```python
N = int(input())

for i in range(1,N+1):
    Space = ' ' * (N-i) # 공백 4, 3, 2, 1
    Star = '*' * (2*i-1) # 별 1, 3, 5, 7, 9
    print(Space + Star)

for i in range(N-1,0,-1): # range(4, 0, -1) -> i = 4, 3, 2, 1
    Space = ' ' * (N-i) # 공백 1, 2, 3, 4
    Star = '*' * (2*i-1) # 별 7, 5, 3, 1b
    print(Space + Star)
```

한꺼번에 해도 되지만 조금 번잡해보일 듯해서 공백과 별을 나눠서 정의하고 print하였다.
