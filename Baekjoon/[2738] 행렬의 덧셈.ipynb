```python
N, M = map(int, input().split())

a = []
for i in range(N):
    row = list(map(int, input().split()))
    a.append(row)
    
b = []
for i in range(N):
    row = list(map(int, input().split()))
    b.append(row)

c = []    
for i in range(N):
    row = []
    for j in range(M):
        row.append(a[i][j] + b[i][j])
    c.append(row)
    
for i in c:
    print(' '.join(map(str, i)))
```
