예제1: 36진법 ZZZZZ
- Z = 35 라고 문제에 주어져있으므로 출력값은 아래와 같다

```python
(35*(36**4))+(35*(36**3))+(35*(36**2))+(35*(36**1))+ (35*(36**0))
```

A=10, B=11, ... Z=35 -> 각 알파벳 대문자에 할당된 숫자 어떻게 나타내지?
- 아스키코드 활용 // ord() : A의 아스키코드는 '65'이다.

```python
N = input()
B = int(input())

for i in range(len(N)):
    print((ord(N[i])-55) * (B**(len(N)-i-1)))
```

```python
N = input()
B = int(input())

answer = 0
for i in range(len(N)):
    answer += (ord(N[i])-55) * (B**(len(N)-i-1))

print(answer)
```

왜 백준에서는 런타임에러가 뜨는거지...했는데 1~9 인경우 생각안해서인 듯해서 다시 작성!

```python
N = input()
B = int(input())

answer = 0
for i in range(len(N)):
    if type(N[i]) == int:
        answer += int(N[i]) * (B**(len(N)-i-1))
    else:
        answer += (ord(N[i])-55) * (B**(len(N)-i-1))

print(answer)
```

```python
N = input()
B = int(input())

answer = 0
for i in range(len(N)):
    if type(N[i]) == int:
        answer += int(N[i]) * (B**(len(N)-i-1))
    elif type(N[i]) != int:
        answer += (ord(N[i])-55) * (B**(len(N)-i-1))

print(answer)
```

계속해서 런타임 에러 발생..🥹🥹

```python
N, B = input().split()
B = int(B)
answer = 0

for i, j in enumerate(N):
    try:
        if int(j):
            answer += int(j) * B ** (len(N)-i-1)
    except:
        answer += (ord(j)-55) * B ** (len(N)-i-1)
        
print(answer)
```

이렇게 변경해서 런타임에러는 겨우 고쳤는데... 이유는 모르겠다ㅠㅠ int 1~9 신경안쓴 거는 내가 놓친 부분이 맞긴하지만, 위의 코드에서
- N = input()
- B = int(input())
- 이처럼 따로 입력하면 또 런타임 에러 발생... 정확한 이유를 알려줬으면 좋겠다
