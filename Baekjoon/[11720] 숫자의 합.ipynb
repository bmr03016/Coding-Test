```python
N = int(input())
word = str(input())
answer = 0

for w in word:
    answer += int(w)

print(answer)
```

- 일단, 밑에 입력되는 문자열이 N자리의 수인지 우선적으로 신경쓰지는 않았다... (이것때문에 계속 꼬아서 생각하게 되는 것 같아서)
- 원래는 for i in range(len(word)): answer += int(word[i])로 작성했었는데 위의 식이 좀 더 간단한 듯하여 변경하였다.

```python
N = int(input())
word = str(input())
answer = 0

for i in range(len(word)):
    answer += int(word[i])

print(answer)
```

이 식도 가능하다🙆
