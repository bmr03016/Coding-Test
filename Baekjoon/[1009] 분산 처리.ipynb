근데 해당 코드는 시간 초과 뜬다...

```python
t = int(input())
data = [list(map(int, input().split())) for _ in range(t)]
result = []

for i in range(t):
    if data[i][0] % 10 == 0: # 10으로 나누어질 때
        result.append(10) # 데이터 처리하는 컴퓨터는 10번 컴퓨터
    else: 
        result.append(pow(data[i][0],data[i][1], 10)) # 앞서 코드의 pow(a,b) % 10
    
    # result = [1,7,6,1,9]

for j in result:
    print(j)
```

시간초과 해결했다...!!

## 🌟 pow(base, exp, mod) / pow(base, exp)

- 앞의 값은 base^exp % mod 값을 말한다.
- 즉, pow(a,b,c) = a**b%c
- pow(base, exp) = 제곱 연산함수 = base ** exp
