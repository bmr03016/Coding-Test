```python
N = int(input()) # N킬로 그램
cnt = 0 # 봉지의 개수

while N >= 0:
    if (N % 5) == 0: # 5의 배수라면
        cnt += (N//5)
        print(cnt)
        break
    N -= 3
    cnt += 1
else: 
    print(-1)
```

- N이 0이상이면 계속 반복
- N이 5로 바로 나누어 떨어진다면
- 몫만큼 cnt(봉지의 개수)를 더해준다.
- 나머지 킬로그램이 남아있지 않으니 바로 cnt를 출력해주고 나와주기
- 3kg의 경우는 N이 3만큼 감소할 때마다 봉지의 개수 하나를 더해준다.
- else: 정확하게 N 킬로그램을 만들 수 없다면 -1을 출력한다

해당 문제가 그리디 알고리즘 문제라는 것은 간파했지만, 3의 배수라는 것을 위처럼 써도 되는지는 생각치 못했다...

```python
N = int(input()) # N킬로 그램
cnt = 0 # 봉지의 개수

while N >= 0:
    if (N % 5) == 0: # 5의 배수라면
        cnt += (N//5)
        print(cnt)
        break
    N -= 3
    cnt += 1
    
    if N < 0:
        print(-1)
```

'N에 3을 계속해서 빼고 음수가 된다면 -1를 print하라'가 오히려 나에게 이해하기 더 편하지 않을까 싶다
