위의 소스를 보면 n=0, n=1인 경우를 제외하고 모든 정수가 fibonacci(n-1) + fibonacci(n-2)를 return한다. 우리가 원하는 값은 0의 개수와 1의 개수니 이를 규칙으로 만들어보면 해당 표가 나온다.

```python
from IPython.display import Image
Image('./images/피보나치규칙.png')
```

```python
T = int(input()) # 테스트의 개수

for i in range(T):
    n = int(input())
    zero = 1 # 0이 출력되는 횟수 (n=0일때, zero=1이되니 초기값을 1로 설정해둠)
    one = 0 # 1이 출력되는 횟수
    tmp = 0 # 임시 저장소
    
    for _ in range(n):
        tmp = one # tmp = 0
                  # tmp = 1
                  # tmp = 1
        one = one + zero # one = 0 + 1 = 1
                         # one = 1 + 0 = 1
                         # one = 1 + 1 = 2
        zero = tmp # zero = 0 -> n = 1일때의 값 1 0
                   # zero = 1 -> n = 2일떄의 값 1 1
                   # zero = 1 -> n = 3일때의 값 1 2
    print(zero, one)
```

n이 3일 경우 3번 반복하여 값을 구한다.
- tmp 임시저장소에 one의 값을 잠시 저장해두고
- one은 one + zero 값을 더해준 값으로 초기화하낟.
- 마지막으로 zero는 잠시 저장해둔 tmp값으로 초기화
