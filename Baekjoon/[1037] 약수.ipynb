예제 입력 3을 보고 약수의 수들을 오름차순으로 정렬하고 제일 값이 작은 약수와 큰 함수의 곱이 출력값이라는 것을 알아차렸다. 조금 더 까다로운 예제 4를 통해도 확인해보면,

```python
lst = [14, 26456, 2, 28, 13228, 3307, 7 ,23149, 8, 6614, 46298, 56, 4, 92596]
lst.sort()
print(lst)
print(lst[0] * lst[-1])
```

출력값이 위의 이미지와 동일하다는 걸 알 수 있다. 하지만 위의 식은 예제 2처럼 약수가 1개인 경우에는 해당하지 않으므로 if문을 사용해서 구별해주었다😊

```python
n = int(input())
m = list(map(int, input().split()))

m.sort()

if n != 0:
    print(m[0]*m[-1])
else:
    print(m[0]*2)
```

파이썬 내장함수인 max(), min()을 사용해서도 가능하다.
