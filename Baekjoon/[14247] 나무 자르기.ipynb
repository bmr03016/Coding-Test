나무들이 자라는 길이 2,7,3,4,1 중 큰 값일 수록 나중에 가져가는게 유리 

```python
n = int(input())
H = list(map(int, input().split())) # 처음 나무의 길이
A = list(map(int, input().split())) # 나무들이 자라는 길이
answer = 0
day = 1

while day <= n:
    i = A.index(min(A)) 
    answer += (A[i] * (day-1)) + H[i]  
    H[i], A[i] = 0, max(A)+1 # min에 안걸리게 임의의 값 max(A)+1로 변경
    day += 1
    
print(answer)
```

- while 문을 통해 day <= 5인 경우 아래의 식이 반복되도록 했다.
- 자라는 길이 * (day-1) + 기존 나무의 길이를 계속해서 더해주고
- H[i], A[i]를 각각 0, max(A)+1로 바꾸어 주었다. (처음에는 remove함수를 썼는데 그러면 원소의 삭제때문에 리스트가 out of range 되게 되어 해당 값으로 대체되게 변경했다.)
- 이 때, A[i]를 max(A)+1 값으로 쓴 이유는 0으로 바꿀 시에 0이 min이 되기 때문에 큰 임의의 숫자가 필요하다고 생각했다. 근데 값들이 뭐가 들어올 지 모르니까 리스트에서 가장 큰 숫자이게 바꾸고자 max(A)+1 을 설정하게 되었다.
- 그리고 다음날이 되니 + 1, day가 5일 때까지 반복
- 😅: 처음에는 예제를 대입해서 생각하다보니 'day <= 5'로 작성했었는데, 그 때문에 풀이가 틀렸다고 나와서 당황했었다...

```python
n = int(input())
H = list(map(int, input().split()))  # 처음 나무의 길이
A = list(map(int, input().split()))  # 나무들이 자라는 길이
answer = 0
day = n

while day >= 1:
    i = A.index(max(A))
    answer = answer + (A[i] * (day - 1)) + H[i]
    H[i], A[i] = 0, 0
    day -= 1

print(answer)
```

- 생각하다보니 **그리디 알고리즘**방식으로 큰 값인 경우를 먼저 생각해도 됐다.
- 이럴 경우, H[i], A[i] 모두 0의 값으로 후에 대체되는 점 때문에 좀 더 간단해지는 점?
- 반대로 day는 계속해서 빼주고 day 가 1이상일 때만 루프문을 돌리게 하면 된다.

```python
n = int(input())
H = list(map(int, input().split())) 
A = list(map(int, input().split()))

lst = [[A[i], H[i]] for i in range(n)]
lst.sort()
answer = 0

for i in range(n):
    answer += lst[i][1] + lst[i][0] * i
    
print(answer)
```
