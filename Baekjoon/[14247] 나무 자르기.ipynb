```python
n = int(input())
H = list(map(int, input().split())) #   
A = list(map(int, input().split())) #   
answer = 0
day = 1

while day <= n:
    i = A.index(min(A)) 
    answer += (A[i] * (day-1)) + H[i]  
    H[i], A[i] = 0, max(A)+1 # min    max(A)+1 
    day += 1
    
print(answer)
```

```python
n = int(input())
H = list(map(int, input().split()))  #   
A = list(map(int, input().split()))  #   
answer = 0
day = n

while day >= 1:
    i = A.index(max(A))
    answer = answer + (A[i] * (day - 1)) + H[i]
    H[i], A[i] = 0, 0
    day -= 1

print(answer)
```

```python
n = int(input())
H = list(map(int, input().split())) 
A = list(map(int, input().split()))

lst = [[A[i], H[i]] for i in range(n)]
lst.sort()
answer = 0

for i in range(n):
    answer += lst[i][1] + lst[i][0] * i
    
print(answer)
```
