🌟 **규칙 발견**
- 각 선의 색상이 하나의 layer이라 치면 해당 배열의 개수에서 규칙이 발견됨
- 1 -> 6 -> 12 -> 18 
- 1 시작점을 기준으로 6의 배수로 증가함

```python
N = int(input())

max_num, layer = 1, 1

while max_num < N: # 1 < 13 
    max_num += 6*layer # (6*1) + 1 -= 7 > (6*2) + 7 = 19 -> (6*3) + 19 = 37
    layer += 1

print(layer) # 13은 layer = 3에 있다 = 3개 방을 지나간다
```
