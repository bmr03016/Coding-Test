```python
N = int(input())

max_num, layer = 1, 1

while max_num < N: # 1 < 13 
    max_num += 6*layer # (6*1) + 1 -= 7 > (6*2) + 7 = 19 -> (6*3) + 19 = 37
    layer += 1

print(layer) # 13 layer = 3  = 3  
```
