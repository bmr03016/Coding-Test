## 🌟 Right Answer

```python
lst = []

for _ in range(5):
    lst.append(input())
    #print(lst)

for i in range(15): # 한 줄에 최대 15글자 -> 15번 반복
    for j in range(len(lst)):
        if i < len(lst[j]):
            print(lst[j][i], end ="")
```

확인을 위해 print(lst)를 해보았음/ 해당 부분 주석 처리하면 맨 아래만 출력됨
- lst의 j번째 인덱스에 있는(ABCDE)에 i번째 원소(i=0이면 A) lst[0][0] = A, lst[0][1] = a 순서대로 출력
- Aa0FfB... 처럼 공백없이 출력되기 위해 end = "" 추가해줌
- end = ""를 해주지 않게 되면 아래처럼 계속해서 개행되서 출력이 됨
