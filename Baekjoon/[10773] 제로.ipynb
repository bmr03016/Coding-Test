```python
K = int(input())
lst = []
answer = []

for i in range(K): # i = 0, 1, 2, 3 
    lst.append(int(input())) # [3, 0, 4, 0]
    
    if lst[i] != 0: # 0이 아니면
        answer.append(lst[i]) # 해당 값을 추가해주고
    else: # 0이면
        answer.pop() # 앞의 값을 삭제한다(마지막 값이기에 pop사용)

print(sum(answer))
```

```python
K = int(input())
lst = []
answer = []

for i in range(K): # i = 0, 1, 2, 3 
    lst.append(int(input())) # [3, 0, 4, 0]
    
    if lst[i] != 0: # 0이 아니면
        answer.append(lst[i]) # 해당 값을 추가해주고
    else: # 0이면
        answer.pop() # 앞의 값을 삭제한다(마지막 값이기에 pop사용)

print(sum(answer))
```

예제 1을 입력해도 동일한 출력값이 나온다. 
- 초반에는 sum 내장함수를 사용할 것이라 0으로 바꿔도 상관없다고 생각했으나, 예제2처럼 0이 여러 번 나오는 경우를 생각하지 못했다.
- 그래서 새로운 리스트를 생성해서 i때문에 list our of range 오류가 나지 않도록 해주었다.
- 맨 처음의 답안과 달리 '0'과 같은 원소가 없기 때문에 맨 마지막 원소를 제거해주면 되서 pop을 사용하였다.

```python
K = int(input())
lst = []

for i in range(K): # i = 0, 1, 2, 3 
    i = int(input())
    
    if i == 0: 
        lst.pop()
    else:
        lst.append(i)

print(sum(lst))
```
