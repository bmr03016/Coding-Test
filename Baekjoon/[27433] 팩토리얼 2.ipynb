```python
N = int(input())

fac = 1
for i in range(1, N+1):
    fac *= i
print(fac)
```

### 팩토리얼 재귀함수

```python
def fac(n):
    if n > 1:
        return n*fac(n-1)
    else:
        return 1

n = int(input())
print(fac(n))
```
