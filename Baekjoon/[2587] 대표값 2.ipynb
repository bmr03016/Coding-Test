```python
lst = []

for i in range(5):
    i = int(input())
    lst.append(i)

lst.sort()
print(int(sum(lst)/5))
print(lst[2])
```

처음에는 8번째 줄에 있는 부분을 단순히 sum(lst)/5 로 출력했는데 그러면 34.0으로 출력이되더라. 그래서 정수형 int로 감싸줌

```python
lst = []

for i in range(5):
    i = int(input())
    lst.append(i)

lst.sort()
print(sum(lst)//5)
print(lst[2])
```
