```python
N, M = map(int, input().split())
arr = [i for i in range(1,N+1)]

for _ in range(M): # M번 반복
    i,j = map(int, input().split())
    arr[i-1:j] = arr[i-1:j][::-1]

print(*arr)
```

오히려 reverse 함수쓰는게 좀 더 복잡할 듯해서 역순을 만들기 위해 [::-1] 를 썼다.

```python
N, M = map(int, input().split())
arr = [i for i in range(1,N+1)]

for _ in range(M): 
    i,j = map(int, input().split())
    temp = arr[i-1:j]
    temp.reverse()
    arr[i-1:j] = temp

print(*arr)
```

위처럼 temp를 통해 임시로 저장하고 reverse한 다음 다시 바꿔주는 형식도 가능하나, reverse() 함수를 쓰게되면 시간이 조금 더 걸리는 듯 하다.
