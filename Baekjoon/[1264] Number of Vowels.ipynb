# 모음의 개수

```python
from IPython.display import Image
Image('images/vowel.png')
```

```python
vowel = ['a', 'e', 'i', 'o', 'u']

while True:
    cnt = 0  #(1)
    s = list(input().lower())  #(2)
    
    if s[0] == '#':  #(3)
        break
        
    for i in s:  #(4)
        if i in vowel:
            cnt += 1
            
    print(cnt)
```

- (1) 모음 개수 세는 cnt
- (2) 문자열을 리스트로 입력받되, 소문자로 다 변경시켜줌
- (3) '#' 한 글자를 입력받게 되면 종료! (한 글자만 주어진다고 해서 첫 글자가 '#'인 경우만 가정하였다)
- (4) s의 문자열에 모음이 존재할 경우 cnt += 1 (for 반복문)
