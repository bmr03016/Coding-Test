```python
table = [list(map(int, input().split())) for _ in range(9)]

max_num = 0
max_row, max_col = 0, 0 # 최댓값이 위치하는 행 번호, 열 번호
for row in range(9): # 행
    for col in range(9): # 열
        if max_num <= table[row][col]: # 최댓값보다 작을 때
            max_row = row + 1 # 행을 더해준다
            max_col = col + 1 # 열을 더해준다
            max_num = table[row][col]

print(max_num)
print(max_row, max_col)
```
