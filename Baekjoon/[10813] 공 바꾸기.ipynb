````python
N, M = map(int, input().split())
arr = [i for i in range(1,N+1)]

for _ in range(M): # M 
    i,j = map(int, input().split())
    arr[i-1], arr[j-1] = arr[j-1], arr[i-1]
    print(*arr)

print(*arr) # print(arr) -> [3, 1, 4, 2, 5]
````
