```python
while True: # (1)
    try:
        print(input())
    except EOFError: #(2)
        break
```

- (2) 입력값이 들어오지 않는다면(EOFError:End of file 상태) break를 걸어줌

While 구문에서 try, except를 사용하는 건 조금 익숙해졌지만, EOFError는 처음 보는 거라 기록해보았다...!
