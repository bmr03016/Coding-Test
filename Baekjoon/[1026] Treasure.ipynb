```python
N = int(input())
A = list(map(int,input().split()))
B = list(map(int,input().split()))

A.sort()
B.sort(reverse=True)
S = 0
for i in range(N):
    S += A[i] * B[i]

print(S)
```

- S의 값을 가장 작게 만들기 위해(최솟값)을 구하는 문제라는 부분에서 그리디 알고리즘이라는 힌트를 찾을 수 있었다.
- 그래서, 가장 큰 단위부터 값을 처리하자!는 생각을 했고,
- 첫번째 예제 기준 A는 0, 1, 1, 1, 6으로 배열(오름차순)
- B는 8, 7, 3, 2, 1로 배열(내림차순)으로 하면 S값을 제일 작게 만들 수 있을 것이라 생각했다.
- 문제에 주어진 힌트를 통해서 A, B를 리스트 형태로 만들고 sort함수로 각각 오름차순, 내림차순 배열하였다.
