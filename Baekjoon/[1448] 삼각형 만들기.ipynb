삼각형을 만들려면 N개의 빨대 중 고른 3개의 빨대에서 **제일 긴 빨대보다 나머지 빨대 두개의 합이 더 커야한다**
이를 이용하면, 아래와 같다.

```python
for _ in range(int(input())):
    int(input())
```

```python
N = int(input()) # 5
straw = [] 

for i in range(N): # i = 4, 5, 6, 7, 20
    i = int(input()) 
    straw.append(i) # (1) straw라는 list에 입력값을 모두 append 한다

straw.sort(reverse=True) # 큰 값부터 탐색하는게 효율적일 것 같아서 내림차순

for j in range(N-2): # j = 0, 1, 2, 3
                        # straw[j] = 20, 7, 6, 5, 4
    if straw[j] < straw[j+1] + straw[j+2]: # 20 < 7 + 6
        print(straw[j] + straw[j+1] + straw[j+2])
        break # 최댓값에서 break
    else:
        print(-1)
```

해서 다시쓴 코드! 애초에 answer을 -1로 설정하고 '제일 긴 빨대보다 나머지 빨대의 합이 클 경우' 그 합들을 answer로 바꾸는 식으로 코드를 짰다.

```python
N = int(input()) # 5
straw = [] 
answer = -1

for i in range(N): # i = 4, 5, 6, 7, 20
    i = int(input()) 
    straw.append(i) # (1) straw라는 list에 입력값을 모두 append 한다

straw.sort(reverse=True) # 큰 값부터 탐색하는게 효율적일 것 같아서

for j in range(N-2): # j = 0, 1, 2, 3
                        # straw[j] = 20, 7, 6, 5, 4
    if straw[j] < straw[j+1] + straw[j+2]: # 20 < 7 + 6
        answer = (straw[j] + straw[j+1] + straw[j+2])
        break # 최댓값에서 break

print(answer)
```
