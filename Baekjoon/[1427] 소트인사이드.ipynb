```python
N = str(input())
lst = []

for i in range(len(N)):
    lst.append(N[i])

lst.sort(reverse=True)
print(''.join(lst))
```

- index를 활용하기 위해서 N를 int형이 아닌 문자(str)로 입력받았고, 해당 속성때문에 range는 len(N)로 받아야 했다.
- 또한 처음에는 print(*lst)로 하면 된다고 생각했으나, 이렇게 할 시에 4 3 2 1 공백과 함께 출력되므로 해당 부분은 평소에 유의해야하지 싶다.
- 공백없이 출력하기 위해 ''.join 활용하였다.

```python
lst = list(input())
lst.sort(reverse=True)
print(''.join(lst))
```

 📍아니면 그냥 input만 하면 str로 입력받는다고한다. 바로 리스트에 입력받아서 정렬해주면 더 간단히 구현 가능하다.
