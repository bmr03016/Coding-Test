```python
#import sys

n = int(input())
stack = []

for i in range(n):
    command = input().split()
    # command = sys.stdin.readline().split()
    
    if command[0] == 'push':
        stack.append(command[1])
    elif command[0] == 'pop':
        if len(stack) == 0:
            print(-1)
        else:
            print(stack.pop())
    elif command[0] == 'size':
        print(len(stack))
    elif command[0] == 'empty':
        if len(stack)==0:
            print(1)
        else:
            print(0)
    elif command[0] == 'top':
        if len(stack)==0:
            print(-1)
        else:
            print(stack[-1])
```

- 백준에서는 import sys를 하고, sys.stdin.readline().split()를 사용했는데

### sys.stdin.readline 특징
1. 한 줄 단위로 입력받음.
2. 개행문자가 같이 입력됨. (예 - 5 입력 → 5\n 저장)  -> 그래서 속도가 빠르다!
3. 개행문자를 제거하는 과정이 필요
4. str 형태로 저장됨
