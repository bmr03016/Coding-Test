```python
n = int(input())

cnt = [0] * (n+1) # 연산을 사용하는 횟수 최소값
                  # 계산된 값을 저장(정수만큼)

for i in range(2, n+1):
    # -1을 뺀다
    cnt[i] = cnt[i-1] + 1 # 계산 결과가 아닌 횟수를 저장하는 것이기에 +1
    
    # 3으로 나눈다
    if i%3 == 0 :
        cnt[i] = min(cnt[i], cnt[i//3]+1)
    
    # 2으로 나눈다    
    if i%2 == 0:
        cnt[i] = min(cnt[i], cnt[i//2]+1)

print(cnt[n])
```

- 앞서의 세 방법을 다 시도해서 최소값을 구한다.
- if, elif를 처음에 사용했었는데 이 때, 계산되지 않은 부분이 생기는 듯해서 if만 사용하였다.
