# 평균은 넘겠지

```python
from IPython.display import Image
Image('images/average.png')
```

```python
c=int(input())

for i in range(c):  #(1)
    A=list(map(int,input().split()))
    student = 0 
    for j in A[1:]:  #(2)
        avg=sum(A[1:])/A[0] 
        if j>avg:
            student += 1
    rate=student/A[0]*100  #(3)
    print('{0:0.3f}%'.format(rate))
```

(1)
- 해당 반복문은 테스트케이스 개수만큼 반복한다.
- A에 학생의 수와 그 학생들의 점수를 입력받는다.
- 변수 student 는 평균을 넘는 학생의 수를 저장하기 위한 변수이다.

(2) 
- 해당 반복문은 리스트에서 학생의 점수(j)에 접근한다.
- 평균을 구한다음
- 만약 j(학생의 점수)가 평균보다 크면 student를 카운트 해준다.

(3)
- 평균을 넘는 학생들의 비율을 구한다
- 소수점 셋째 자리까지 출력: '{0:0.3f}%'.format

**'{0:0.3f}%'.format 이게 그렇게도 아직 익숙치 않더라. 이번에 잘 기억해둘 것!**
