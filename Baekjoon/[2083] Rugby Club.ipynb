```python
while True:
    name, age, weight = input().strip().split()  #(1)
    
    if name=="#" and age=="0" and weight=="0":  #(2)
        break
    if int(age) > 17 or int(weight) >= 80:  #(3)
        print(f"{name} Senior")
    else:
        print(f"{name} Junior")
```

- (1) 무한루프로 이름, 나이, 무게를 입력 받음
- (2) 입력을 # 0 0 으로 받게 되면 반복문 종료!
- (3) 나이가 17세보다 많거나, 몸무게가 80kg 이상이면 Senior, 그렇지 않으면 Junior
