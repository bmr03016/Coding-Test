```python
A, B = list(input().split())      
A = list(A)                                                     
B = list(B)

A.reverse()
B.reverse()

if A >= B:
    print(*A, sep='')
else:
    print(*B, sep='')
```

```python
A, B = list(input().split())      
A = list(A[::-1])                                                     
B = list(B[::-1])

print(*max(A,B), sep='')
```

- 역순은 reverse함수도 가능하지만, [::-1]으로도 가능!
- 그리고 아래와 같이 입력을 문자열로 받고 [::-1] 로 역순처리 및 int형으로 변경하면 좀 더 간단하게 풀이식을 쓸 수 있다.
- 🥹 reverse를 활용해야 겠다는 생각에 이렇게 풀이할 수도 있다는 점을 놓쳤다...

```python
A, B = input().split()
 
A = int(A[::-1])
B = int(B[::-1])
 
print(max(A, B))
```
