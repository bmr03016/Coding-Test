```python
n, k = map(int, input().split())
answer = 1

for i in range(n, k+1, -1):
    answer *= i
    
print(answer)
```

```python
n, k = map(int, input().split())
answer = 1

for i in range(n, k+1, -1):
    answer *= i
for j in range(k, 0, -1):
    answer //= j
    
print(answer)
```

틀렸다고 나와서 조금 더 고민해보니 k+1 범위가 다른 에제에서는 적용되지 않을 것을 알게 되었다.

```python
n, k = map(int, input().split())
answer = 1

for i in range(n, n-k, -1):
    answer *= i
for j in range(k, 0, -1):
    answer //= j
    
print(answer)
```

k+1을 n-k로 변환. 성공! 😊

```python
def Fac(a):
    if a == 0:
        return 1
    else:
        return a*Fac(a-1)

n,k = map(int, input().split())
print(Fac(n)//(Fac(n-k) * Fac(k)))
```
