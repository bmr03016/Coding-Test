```python
dial = ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQRS', 'TUV', 'WXYZ']
word = input()
time = 0 # 다이얼을 걸기 위해서 필요한 시간

for i in range(len(word)): # range(2) / i = 0, 1
    for j in dial: # j = 'ABC', 'DEF', ...
        if word[i] in j: #word[0] = W in 'JKL'
            time += dial.index(j) + 3 # 리스트.index(원소)
                                    # 'ABC'는 3초 소요, index 1이 커질때마다 시간도 1씩 커짐
    
print(time)    
```

- 'W', 'A' 알파벳이 dial 어디 위치(index)에 있는지 찾고, 해당 값 + 3을 하면 답이 나온다는 것을 생각했는데,
- 막상 식으로 적으려하니 조금 헷갈렸다.
- word[i]가 dial원소에 있으면 답 출력
- j를 dial 원소라고 하자
- i가 필요하니까 for문 len(word)만큼 돌리면 되겠다.
- 라고 거꾸로 좀 풀어서 생각해서 작성...
