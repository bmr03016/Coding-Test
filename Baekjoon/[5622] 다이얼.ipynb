```python
dial = ['ABC', 'DEF', 'GHI', 'JKL', 'MNO', 'PQRS', 'TUV', 'WXYZ']
word = input()
time = 0 #     

for i in range(len(word)): # range(2) / i = 0, 1
    for j in dial: # j = 'ABC', 'DEF', ...
        if word[i] in j: #word[0] = W in 'JKL'
            time += dial.index(j) + 3 # .index()
                                    # 'ABC' 3 , index 1   1 
    
print(time)    
```
