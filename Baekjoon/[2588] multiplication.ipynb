```python
A = int(input())
B = input()

print(A * int(B[2]))
print(A * int(B[1]))
print(A * int(B[0]))
print(A * int(B))
```

- B를 정수로 변환하지 않고 문자열을 활용하였다.
- 슬라이싱을 이용해 각 숫자를 가져오고, 출력할때는 정수로 변환

```python
A = int(input())
B = input()

for i in range(3):
    print(A*int(B[i-1]))
print(A*int(B))
```

- for문을 활용하면 조금 더 간단해진다
- 슬라이싱 i-1로 해주는 거 잊지 말기!
