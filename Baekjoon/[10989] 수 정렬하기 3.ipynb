```python
N = int(input()) # 10
lst = []

for i in range(N):
    i = int(input())
    lst.append(i)
    lst.sort()

for j in lst:
    print(j)
```

- 파이썬 내장 sort함수는 병합정렬로 구현되었기 때문에 O(nlogn)의 시간복잡도를 가져서, 해당 문제같이 input 최댓값이 10,000,000이면 시간초과가 날 수 밖에 없다. 
- 이럴 때 사용하는 것이 **계수정렬**이다.

# 👆 **계수정렬**
- 배열의 인덱스가 sequencial 하게 정렬되어 있다는 특성을 이용한 정렬로, 인덱스를 해당 숫자값으로 사용해서 중복되는 숫자도 카운팅할 수 있다.
- 수의 범위는 작지만, 연산 횟수는 많을 때 사용한다.

```python
N = int(input()) 
lst =[0]*10001 # 입력으로 받을 수 있는 10,000개의 수를 담을 수 있는 배열(10,001)을 만들고

for _ in range(N):
    x = int(input()) # 입력받을 때마다
    lst[x] += 1 # 그 수에 해당하는 인덱스에 +1을 해줘서 값으로 그 수의 개수를 담는다.

for i in range(10001): # 그리고 다시 배열을 돌리면서
    if lst[i] != 0: # 값이 0이 아니라면
        for j in range(lst[i]): # 값만큼 인덱스에 해당하는 숫자를 출력 
            print(i) 
```
