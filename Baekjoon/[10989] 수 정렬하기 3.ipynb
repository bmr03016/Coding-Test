# üëÜ **Í≥ÑÏàòÏ†ïÎ†¨**
- Î∞∞Ïó¥Ïùò Ïù∏Îç±Ïä§Í∞Ä sequencial ÌïòÍ≤å Ï†ïÎ†¨ÎêòÏñ¥ ÏûàÎã§Îäî ÌäπÏÑ±ÏùÑ Ïù¥Ïö©Ìïú Ï†ïÎ†¨Î°ú, Ïù∏Îç±Ïä§Î•º Ìï¥Îãπ Ïà´ÏûêÍ∞íÏúºÎ°ú ÏÇ¨Ïö©Ìï¥ÏÑú Ï§ëÎ≥µÎêòÎäî Ïà´ÏûêÎèÑ Ïπ¥Ïö¥ÌåÖÌï† Ïàò ÏûàÎã§.
- ÏàòÏùò Î≤îÏúÑÎäî ÏûëÏßÄÎßå, Ïó∞ÏÇ∞ ÌöüÏàòÎäî ÎßéÏùÑ Îïå ÏÇ¨Ïö©ÌïúÎã§.

```python
N = int(input()) 
lst =[0]*10001 # ÏûÖÎ†•ÏúºÎ°ú Î∞õÏùÑ Ïàò ÏûàÎäî 10,000Í∞úÏùò ÏàòÎ•º Îã¥ÏùÑ Ïàò ÏûàÎäî Î∞∞Ïó¥(10,001)ÏùÑ ÎßåÎì§Í≥†

for _ in range(N):
    x = int(input()) # ÏûÖÎ†•Î∞õÏùÑ ÎïåÎßàÎã§
    lst[x] += 1 # Í∑∏ ÏàòÏóê Ìï¥ÎãπÌïòÎäî Ïù∏Îç±Ïä§Ïóê +1ÏùÑ Ìï¥Ï§òÏÑú Í∞íÏúºÎ°ú Í∑∏ ÏàòÏùò Í∞úÏàòÎ•º Îã¥ÎäîÎã§.

for i in range(10001): # Í∑∏Î¶¨Í≥† Îã§Ïãú Î∞∞Ïó¥ÏùÑ ÎèåÎ¶¨Î©¥ÏÑú
    if lst[i] != 0: # Í∞íÏù¥ 0Ïù¥ ÏïÑÎãàÎùºÎ©¥
        for j in range(lst[i]): # Í∞íÎßåÌÅº Ïù∏Îç±Ïä§Ïóê Ìï¥ÎãπÌïòÎäî Ïà´ÏûêÎ•º Ï∂úÎ†• 
            print(i) 
```
