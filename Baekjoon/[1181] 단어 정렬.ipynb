```python
N = int(input())
lst = []

for _ in range(N): # N개의 단어가 들어온다
    lst.append(input())

lst = list(set(lst)) # 중복 없애기 위한 set '{}' 그리고 list로 다시 감싸줘야 함
lst.sort() # 사전순 정렬
lst.sort(key=lambda x:len(x)) # 길이순 정렬

for i in lst:
    print(i)
```

해당 식이 좀 헷갈리는거 같아서 다시 정리

```python
N = int(input())
word = set() # 중복을 없애기 위해 set로 입력받는다.

for _ in range(N): # N개의 단어가 들어온다
    word.add(input()) # 세트는 append를 쓰지못하고 add로

lst = list(word) # 정렬을 위해 list로 바꿔주기
lst.sort(key=lambda x:(len(x),x)) # 길이순, 사전순 정렬

for i in lst:
    print(i)
```

- 아예 단어를 set로 입력받았고 (중복을 없애기 위함)
- set는 append를 쓰지못하므로 add를 활용하여 N개의 단어를 입력한다.
- 정렬을 하기 위해 word를 list형식으로 바꿔주고
- lambda식을 활용하여(key) 길이순, 사전순으로 정렬해준다. 이때 두 개의 기준으로 정렬하기 위해 () 괄호로 한번 더 감싸줘야 한다.
- 그리고 lst에 있는 원소들을 하나하나 출력해준다.
