- B[k]가 몇 번째 숫자인지 알기 위해
- B[k]보다 작은 숫자의 개수를 찾기로 했다

```python
N = int(input())
k = int(input())

start, end = 0, k

while start <= end:
    mid = (start+end)//2
    cnt = 0

    for i in range(1, N+1):
        cnt += min(mid//i, N)

    if cnt >= k:
        end = mid - 1
    else:
        start = mid + 1

print(start)
```

- 모든 배열을 검색하지 않고, k를 end 값으로 설정하였다.
- 이 사이 값(0~k)에서 **이진 검색법**을 활용했다.
- 'mid//i ->  mid보다 작거나 같은 숫자의 개수' or 최대값 N을 반환하게 코드를 짰다.
- print(start)를 썼지만 print(mid) 또한 가능하다. 하지만 end는 불가능
