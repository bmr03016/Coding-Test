### 이진 검색 알고리즘을 활용하여 문제에 접근하였다.

```python
Image('images/budget_explanation.png', width=600)
```

```python
# 입력
n = int(input()) # 지방의 수: 4
requests = list(map(int, input().split())) # [120, 110, 140, 150]
m = int(input()) # 예산

low, high = 0, max(requests)

# 이분 탐색
while low <= high:
    mid = (low + high)//2
    upper_bound = 0

    for i in requests:  # for 120 in [120, 100, 140, 150]
        upper_bound += min(mid, i)

    if upper_bound <= m:
        low = mid + 1
    else:
        high = mid - 1
print(high)
```

- 분명히 주피터 노트북에서는 잘되는데 백준에서는 런타임 에러가 발생해서,, 오류 수정하느라 애먹었다ㅠㅠ

```python
n = int(input()) 
requests = list(map(int, input().split()))
m = int(input()) 
requests.sort()

low, high = 0, max(requests)

while low <= high:
    mid = (low + high)//2
    upper_bound = 0

    for i in requests:
        upper_bound += min(mid, i)

    if upper_bound <= m:
        low = mid + 1
    else:
        high = mid - 1
print(high)
```

- sort 함수로 정렬했더니 백준에서도 에러없이 성공! (sort함수 안쓴 정답들도 봤는데,, 왜 오류떴는지는 모르겠다..)
