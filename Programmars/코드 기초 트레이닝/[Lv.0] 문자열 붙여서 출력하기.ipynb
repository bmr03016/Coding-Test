붙여서 출력한다고 해서(공백제거) replace(" ", "")로 사용하려 했는데

```python
str1, str2 = input().strip().split(' ')
print(str1+str2)
```

프로그래머스에 첫줄이 주어져 있었기에 단순히 서로의 문자열을 붙여주는게 더 간단하다고 생각하였다

```python
print(input().strip().replace(' ', ''))
```

하지만 str1, str2 변수를 없애버리고 앞서의 함수 replace를 써서 더 간단하게 식을 쓸 수도 있다.
