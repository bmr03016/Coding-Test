```python
def solution(n, control):
    answer = n
    for i in range(len(control)):
        if control[i] == 'w':
            answer += 1
        elif control[i] == 's':
            answer -= 1
        elif control[i] == 'd':
            answer += 10
        elif control[i] == 'a':
            answer -= 10
    return answer
```

- 처음에는 아래와 같이 작성했는데, 계속해서 실패떠서 조금 더 깊이 생각해보자해서 작성한 위의 코드..
- 근데 아래 코드에서 맨 처음에 n=0이라고 정의하고 들어가서 계속 에러가 뜬거였다.. 

```python
def solution(n, control):
    for i in control:
        if i == 'w':
            n += 1
        elif i == 's':
            n -= 1
        elif i == 'd':
            n += 10
        elif i == 'a':
            n -= 10
    return n
```

📍 그리고 다른 분의 코드보게 되었는데, 한번 짚고 넘어가면 좋을 거 같아서 리뷰!

```python
def solution(n, control):
    key = dict(zip(['w','s','d','a'], [1,-1,10,-10]))
    return n + sum([key[c] for c in control])
```

1. 딕셔너리 = dict(키1=값1, 키2=값2)
2. 딕셔너리 = dict(zip([키1, 키2], [값1, 값2]))
3. 딕셔너리 = dict([(키1, 값1), (키2, 값2)])
4. 딕셔너리 = dict({키1: 값1, 키2: 값2})

두번째 방법 / 키 리스트와 값 리스트를 묶어 활용
- key[c]는 w, s, d, a에 해당하는 1,-1,10,-10의 값을 말한다
