```python
def solution(a, b, c):
    if a == b and b == c and c == a:
        # 세 숫자가 모두 같을 때
        return (a + b + c) * (a**2 + b**2 + c**2) * (a**3 + b**3 + c**3)
    elif a != b and b!= c and c != a:
        # 세 숫자가 모두 다를 때
        return a + b + c
    else:
        # 세 숫자가 어느 두 숫자는 같고 나머지 다른 숫자는 다를 때
        return (a + b + c) * (a**2 + b**2 + c**2)
```

세 숫자가 어느 두 숫자는 같고 나머지 다른 숫자는 다를 때를 식으로 풀이해봤을 때 계속 오류가 떠서 아예 else로 빠지게 하는 식을 했더니 정답...! 😊 근데 내가 식을 잘못적은 거였고, 'a==b or a==c or b==c'으로 조건문 설정 했으면 됐던 거였는데 너무 복잡하게 생각했던 듯 하다 ㅎㅎ
