```python
import numpy as np

def solution(num_list):
    if np.prod(num_list) < np.sum(num_list)**2: 
        return 1
    return 0
```

원래는 아래와 같이 풀고 있었는데 계속 안되길래 위처럼 numpy를 import해주고 prod 함수를 써서 풀었다.

```python
def solution(num_list):
    a = 1
    b = 0
    for n in num_list:
        a *= n
        b += n
        
    if a < b**2: return 1
    return 0
```

풀고 나서 왜 안됐는지 유심히 다시 봤는데, 안된 이유가 두 가지 있었다.
- 1) 곱을 해야하는데 a를 0으로 정의하고 시작한 것
- 2) if구문을 더 들여쓰기 한 것
