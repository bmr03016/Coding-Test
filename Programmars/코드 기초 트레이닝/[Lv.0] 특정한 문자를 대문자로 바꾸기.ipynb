해당 문제는 두 가지 방법으로 풀이 가능하다.

```python
def solution(my_string, alp):
    answer = ''
    for i in my_string:
        if i == alp:
            answer += i.upper()
        else: 
            answer += i
    return answer
```

```python
def solution(my_string, alp):
    return my_string.replace(alp, alp.upper())
```
