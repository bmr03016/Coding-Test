## 1. for 

```python
def solution(num_list):
    a = 0
    b = 1
    for i in num_list:
        a += int(i)
        b *= int(i)
    return a if len(num_list)>=11 else b
```

## 2. eval, join  

```python
def solution(num_list):
    if len(num_list) >= 11:
        return eval('+'.join(list(map(str, num_list))))
    else:
        return eval('*'.join(list(map(str, num_list))))
```

## 3. reduce 

```python
from functools import reduce
def solution(num_list):
    return sum(num_list) if len(num_list)>=11 else reduce(lambda a,b: a*b, num_list)
```
