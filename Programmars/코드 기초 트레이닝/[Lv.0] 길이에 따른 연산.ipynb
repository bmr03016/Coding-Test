💡 **리스트의 모든 원소의 곱 구하는 법**
- 1. for문 활용
- 2. eval과 join 함수 사용하기
- 3. reduce함수 사용하기

## 1. for문 활용

```python
def solution(num_list):
    a = 0
    b = 1
    for i in num_list:
        a += int(i)
        b *= int(i)
    return a if len(num_list)>=11 else b
```

## 2. eval, join 함수 사용

```python
def solution(num_list):
    if len(num_list) >= 11:
        return eval('+'.join(list(map(str, num_list))))
    else:
        return eval('*'.join(list(map(str, num_list))))
```

**eval 함수**
- 문자열 식을 수행
- 식만 처리하므로 변수 선언 및 할당과 같은 것은 할 수 없다.

**join 함수**
- 문자열을 다룰 때 유용하게 사용할 수 있는 함수로
- 주어진 요소들을 하나의 문자열로 합쳐주는 역할

## 3. reduce함수 사용

```python
from functools import reduce
def solution(num_list):
    return sum(num_list) if len(num_list)>=11 else reduce(lambda a,b: a*b, num_list)
```

- reduce 함수는 반복 가능한 객체
- functools에서 모듈을 불러와야 함
- lambda 매개변수 : 표현식
- reduce(함수, 시퀀스) 이므로
- reduce(lambda x, y: x + y, [0, 1, 2, 3, 4]) 와 같이 활용된다.

📍 eval, join, reduce, lambda 식에 좀 더 익숙해져야겠다.. 바로 생각이 들 수 있도록!!
