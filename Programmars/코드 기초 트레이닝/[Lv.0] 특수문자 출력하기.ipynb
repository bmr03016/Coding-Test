```python
print("!@#$%^&*(\\'\"<>?:;")
```

- 위처럼 \, ', " 기호앞에 역슬래쉬(\)를 한번 더 작성해주는 것도 방법이지만,

```python
print(r'!@#$%^&*(\'"<>?:;')
```

💡 이외에도 문자열 앞에 r을 붙이면 쉽게 문자열 내의 특수문자를 자동으로 인식가능하다!!!
