```python
def solution(n):
    list = []
    for i in range(n+1):
        if n % 2 == 0 and i % 2 == 0:
            list.append(i*i)
        elif n % 2 != 0 and i % 2 != 0:
            list.append(i)
    return sum(list)
```

- n 이하의 홀수/짝수를 저장할 list를 만듬
- n 이하의 수까지 더해야하는 거니까 for문 range(n+1)까지
- n이 짝수인데 i까지 짝수인 경우에만 list에 append(더해줌)
- 홀수인 경우도 이러한 이치
- list 합계 함수 sum을 사용해서 return
