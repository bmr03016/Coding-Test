```python
def solution(binomial):
    a = int(binomial.split()[0]) 
    op = binomial.split()[1]
    b = int(binomial.split()[2])
    if op == "+":
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
```

문제 풀이를 보니, expression 을 문자열로 받아서 실행시켜주는 **eval() 함수**가 존재했다...😥

```python
def solution(binomial):
    return eval(binomial)
```
