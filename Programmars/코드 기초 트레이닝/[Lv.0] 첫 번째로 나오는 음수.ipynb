```python
def solution(num_list):
    answer = -1
    while True:
        for i in range(len(num_list)):
            if num_list[i] < 0:
                answer = i
                break
        
        return answer
```

📌 여기서 point는
- 1. 첫 번째로 나오는 음수의 인덱스만을 return 해야ㅏ기에 while True - break문을 사용하였다는 점
- 2. answer를 처음에 -1로 정의하고 간 점(음수가 없다면 -1를 return해야하기에 아예 이렇게 정의하고, 음수가 있다면 answer를 i로 바꿔줌)

근데 다른 분들의 풀이를 보니, break문을 쓰려한다고 while True를 적진 않아도 되는 듯하다..ㅎㅎ 새로알게된 사실!
