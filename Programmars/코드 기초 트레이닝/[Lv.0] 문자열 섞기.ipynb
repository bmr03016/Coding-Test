```python
def solution(str1, str2):
    answer = ""
    for i in range(len(str1)):
        answer += str1[i] + str2[i]
        
    return answer
```

- 문자열이기에 ""으로 먼저 정의해주고
- len(str1)만큼 반복문
- str[i] + str2[i] = ab인데
- 반복문 통해 계속 더하고 출력
