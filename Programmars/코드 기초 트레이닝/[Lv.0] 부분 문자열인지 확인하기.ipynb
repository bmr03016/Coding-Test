```python
def solution(my_string, target):
    try:
        my_string.index(target)
        return 1
    except ValueError:
        return 0
```

나는 위처럼 try 구문 및 index 메서드 활용하여 풀이하였는데 (index: 파이썬에 내장된 부분 문자열 인덱스 반환하는 역할하는 내장함수) 아래와 같이 보다 간단하게 할 수 있는 방법이 있었다....

```python
def solution(my_string, target):
    if target in my_string: return 1
    return 0
```
