```python
def solution(arr1, arr2):
    arr1_sum, arr2_sum= 0, 0
    
    if len(arr1) < len(arr2): return -1  #(!)
    elif len(arr1) > len(arr2): return 1
    else: #(2)
        for i in range(len(arr1)):  #(3)
            arr1_sum += arr1[i]
        for i in range(len(arr2)):
            arr2_sum += arr2[i]
            
        if arr1_sum < arr2_sum: return -1 #(4)
        elif arr1_sum > arr2_sum: return 1
        else: return 0
```

- (1)      / arr2   -1 return, arr1   1 return
- (2)     
- (3)    arr1_sum, arr2_sum  
- (4)  (arr1_sum, arr2_sum)    -1, 1, 0 return
