```python
def solution(a, b):
    x = int(str(a) + str(b))
    try:
        return max(x, 2*a*b)
    except x == 2*a*b:
        return x
```

나는 위처럼 '단, a⊕b와 2*a*b가 같으면 a⊕b를 return합니다'에 주목해서 예외처리하려고 try - except- 구문을 썼는데,
생각해보니 두 값이 같으니 굳이 예외처리를 할 필요가 없었다...하하

```python
def solution(a, b):
    return max(int(str(a) + str(b)), 2*a*b)
```
