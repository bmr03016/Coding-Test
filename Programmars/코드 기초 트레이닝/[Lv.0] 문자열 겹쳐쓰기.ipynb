### 틀린 답안

```python
def solution(my_string, overwrite_string, s):
    return my_string.replace(my_string[s:len(overwrite_string) + s], overwrite_string)
```

### 🙆 맞는 답안

```python
def solution(my_string, overwrite_string, s):
    return my_string[:s] + overwrite_string + my_string[len(overwrite_string) + s:]
```

- 테스트 데이터를 볼 수 없어서 잘은 모르겠지만, 아마 replace 함수는 범위에 따라 대체되는 문자열이 여러 번 입력되서 오류가 발생하는 것 같다... 
- 좀 더 잘게 범위를 나눠 문자열을 합해주니 정답으로 인정되었음!
