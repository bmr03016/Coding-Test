```python
def solution(arr):
    answer = []
    for i in arr:
        if i >= 50 and i%2 == 0:
            answer.append(i/2)
        elif i < 50 and i%2 != 0:
            answer.append(i*2)
        else:
            answer.append(i)
    return answer
```

- 리스트는 replace를 쓰지 못하기 때문에 새로운 list를 만들어 append하는 것이 나을 것이라 판단
- 50보다 크거나 같은 짝수이면 i/2 append
- 50보다 작은 홀수이면 i*2 append
- **그렇지 않으면 본연의 i append** (풀이과정 초반에 놓쳤던 부분)
