## 처음에 생각했던 방식

```python
def solution(participant, completion):
    for p in participant: # p = leo, kiki, eden
                          # marina, josipa, nikola, vinko, fillpa
        for c in completion: # c = eden, kiki
                             # josipa, fillpa, marina, nikola
            if p != c:
                return p
```

- 참여자(p)와 완주자(c)가 같지 않은(!=)
- 참여자(p)가 return 값이라 생각했다.
- 해당 접근방식은 좋았으나,3개 중 1개 오답이 나왔고 조금 더 디테일하게 생각해보았다.

## 답

```python
def solution(participant, completion):
    participant.sort()
    completion.sort()
    for p, c in zip(participant, completion):
        if p != c:
            return p
    return participant[-1]
```

- 답의 접근 방식은 맞으나, 오류가 떴을 때는 
- sort() 함수를 적용했을 때 해결될 때가 많았다.
- 정렬을 하게 되면 
- participant = ["eden", "kiki", "leo"]
- completion = ["eden", "kiki"] 로 될 것이기 때문에
- participant의 맨 마지막 값(["leo"])을 return 해주면 깔끔하게 답이 나온다.
